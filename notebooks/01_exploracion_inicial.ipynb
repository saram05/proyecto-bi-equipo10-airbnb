Este notebook tiene como objetivo hacer una primera exploración del dataset de Airbnb NYC para confirmar su estructura, tamaño y tipos de variables. No se realiza limpieza todavía, eso se hará en la Fase 3.

```python
# Importamos las librerías necesarias para la exploración
import pandas as pd
import numpy as np

# Configuración para ver todas las columnas
pd.set_option('display.max_columns', None)

# Cargamos el archivo dataset original descargado de Kaggle.
df = pd.read_csv('../data/raw/AB_NYC_2019.csv')

print(f"Dataset cargado correctamente")
print(f"Filas: {df.shape[0]:,}")
print(f"Columnas: {df.shape[1]}")

#Vemos las primeras filas para entender cómo se ve la información.
df.head()

# Información general
# df.info()

# Estadísticas descriptivas de variables numéricas
df.describe()

# Distritos (neighbourhood_group)
print("Distritos:")
print(df['neighbourhood_group'].value_counts())
print()

# Tipos de habitación (room_type)
print("Tipos de habitación:")
print(df['room_type'].value_counts())
print()

# Cantidad de barrios únicos
print(f"Cantidad de barrios únicos: {df['neighbourhood'].nunique()}")


# Valores nulos por columna
nulos = df.isnull().sum()
nulos_porcentaje = (df.isnull().sum() / len(df)) * 100

resumen_nulos = pd.DataFrame({
    'Nulos': nulos,
    '% Nulos': nulos_porcentaje.round(2)
})

# Mostramos solo las columnas que tienen nulos
resumen_nulos[resumen_nulos['Nulos'] > 0].sort_values('Nulos', ascending=False)


## 7. Observaciones iniciales

# - Hay **valores nulos** en las columnas `last_review` y `reviews_per_month` (~20%), lo cual tiene sentido porque corresponden a hospedajes que nunca han recibido reseñas.
# - También hay algunos nulos en `name` y `host_name`, pero son pocos.
# - El dataset contiene **5 distritos** y **3 tipos de habitación**.
# - La variable `price` tiene valores mínimos de 0, lo cual es un indicio de que habrá que filtrar datos inválidos en la fase de limpieza.
# - Existen outliers evidentes en `price` (valores muy altos), que también tendrán que tratarse en la Fase 3.

```
